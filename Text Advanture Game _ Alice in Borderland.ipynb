```python
"""

Date : January 27 2021
@author : Yuan Ning Chiang

"""

# Using input function for players to play and interact with this game. 
# Keep blank spaces after ":" to help users to read easily 
player = input(prompt = "Register here! Sign your name: \n")

print(f"""
    Hi! {player.capitalize() } : 

Welcome to "Alice in Borderland". The background of this game is based on the Japanese 
Netflix series(2020). 
You are in parallel world with your best friend - Masaki. People who are now in this world to 
play the game is because they don't have their personal goals, they don't have contributions of 
the society, or who want to commit suicide.The resource in the world is limited. We cannot have 
thoes kind of people to survive and waste the resource. The purpose of the game is to help you
realize, how important the life is. 
To survive and return to your world, you will have to play the game and win three sadistic
stages.
  
    
    1st Stage: (question) Answer three consecutive questions.
    2nd Stage: (kill) Who do you want to kill?
    3rd Stage: (door) Are you lucky?
    
{'-'* 100}\n
""")
#to exit the game, import exit funtion adter win or lost the game
from sys import exit
#import random function to generate the random variables in the list or
#randomly select the numbers
import random

#user input function, press enter then the game will continue  
input(prompt = "<Press enter to continue>\n")
print(f"""{'-'* 100}\n""")

# 1 st stage
# using a UDF variable to start the first stage 
def question():
    
    #print the input variables with the first characcter with capital,and explain 
    #the rules of this stage  
    print(f"""Hi {player.capitalize()}!
    
Welcome to the first stage of the game - Please answer three questions 
to survive.We all know that 'KNOWLEDGE IS POWER', however, people in parallel world only 
care about themselves, but not the society nor other people.It's sad that 
lots of people from some part of the world are now starving or suffering 
because of the war.They try so hard to survive because they know the life
is valuable. You have no idea about the world but only thinking 
of yourself. So, now you have to answer the questions to go to the next stage,
the questions are about the history, present and the 
future with regards to the world. The answer have to be consecutive correct.   
{'-'* 100}\n
    """)
    
    #use user input function for the question and use can provide the answer 
    answer_q1 = input(prompt = "What is the surname of the current leader of China?\n")
    
    #changing the characters to lower case by using the casefold()
    #in order to compare the user input and the correct answer
    answer_q1 = answer_q1.casefold()
     
    # using conditional function to check if the three answers are all correct, 
    # user fail if the answer is wrong, then will go to the fail() UDF function     
    if answer_q1 == 'xi':
        print("Correct!! Please answer the second question")
        print(f"""{'-'* 100}\n""")
        #the first answer is correct, printing the second question
        answer_q2 = input(prompt = "Which year did Isreal become independent? please enter in number.\n")
        #changing the string to integer by using int() function
        answer_q2 = int(answer_q2)
        
        # Start of the first nested conditional
        if answer_q2 == 1948:
            print("Correct!Please answer the last question")
            print(f"""{'-'* 100}\n""")
            #the second answer is correct, printing the third question
            answer_q3 = input(prompt = "Which CITY will host the 2024 Summer Olympic game? \n")
            #changing the characters to lower case by using the casefold()
            #in order to compare the user input and the correct answer
            answer_q3 = answer_q3.casefold()
            
            
            # Start of the second nested conditional
            if answer_q3 == 'paris':
                #answer all the questions correctly, go to the second stage
                # using UDF variables kill() to go the the next stage
                print("Correct!You will go to the next stage!")
                print(f"""{'-'* 100}\n""")
                kill()
            
            elif answer_q3 != 'paris':
                print("the correct answer is Paris")
                # wrong answer for third question, so fail
                fail()
            
            else:
                print("Something went wrong!!")
                
                #end of the second nested conditional
            
        elif answer_q2 != 1948:
            print("the correct answer is 1948")
            # wrong answer for second question, so fail
            fail()
        
        else:
            print("Something went wrong!!")
            #end of the first nested conditional
            
    elif answer_q1 != 'xi':
        print("the correct answer is Xi")
        # wrong answer for first question, so fail
        fail()
        
    else:
        print("Something went wrong!!")    
    

# 2 nd stage
# using a UDF variable to start the second stage 
def kill():
    #introduction of the second stage
    print(f"""
Congratulation for surviving in the 1st stage, now you are in the second stage - to kill or not.
In this game you and Masaki will have guns in your hands. Both of you have to chose who will 
die, you or Masaki? Players can commit suicide to save others, or kill the others to survive.
At the same time, Masaki will make his choice as well. Who do you want to kill? 
Please make your choice.

    1) Masaki
    2) Myself
{'-'* 100} \n
    """) 
    
    #using user input function for player to answer the question
    choice = input(prompt = "Your answer is : \n")

    #using casefold() function to change the charatcers to lower case    
    choice = choice.casefold()
    
    # creating the list to simulate Masaki's choice
    situation_list = ['at the same time, Masaki kills you.',
                      'at the same time, Masaki commits suicide.']
    # using 'random' function to choose Masaki's choice in the list[]
    killing = random.choice(situation_list)
    
    
    #using conditional function to compare with the input answer
    if choice == '1' in choice or 'masaki' in choice:
        
        #start the nested conditional if you chose to kill Masaki
        if killing == situation_list[0]:
            print(situation_list[0])
            print(f"""
Unfortunately, you and your best friend want to kill each other.
Both of you dead. Being good friend in the next life! \n
{'-'* 100} \n
        """)
            #by using the fail() UDF variable to go to fail and end the game
            fail()
        elif killing == situation_list[1]:
            print(situation_list[1])
            print(f"""
You chose to kill your best friend and your best friend choose to die!
You kill him but he choose to save you!! How lucky you are!
Please go to the last stage!!\n
{'-'* 100} \n
        """)
            #by using the door() UDF variable to go to the third stage as player win
            door()
            
        else:
            print("something went wrong")
            
            #end the nested conditional if you chose to kill Masaki
    elif choice == '2' in choice or 'myself' in choice:
        
         #start the nested conditional if you chose to kill yourself
        if killing == situation_list[0]:
            print(situation_list[0])
            print(f"""
You chose to save your friend but your best friend killed you! you dead!!
Learning how to choose friends is pretty important!!\n
{'-'* 100} \n
        """)
            #by using the fail() UDF variable to go to fail and end the game
            fail()
            
        elif killing == situation_list[1]:
            print(situation_list[1])
            print(f"""
Both of you chose to commit suicide!You and your best friend sacrifice to save 
each other's life!!
Therefore, you and Masaki both earn one more chance to play this game\n
{'-'* 100} \n
        """)
            #by using the kill() UDF variable to go reply in this stage
            kill()
        else:
            print("something went wrong")
            #end the nested conditional if you chose to kill yourself
    else:
        print("something went wrong")
        
# 3 rd stage
# using a UDF variable to start the third stage 
def door():
    #introduction of the third stage 
    print(f"""
Congratulation for survive in the last 2 stages, you are in the last stage.
Luck is also part of the strength! In this game, there are 3 doors in front of you.
2 of the doors are the exits of the game, however, behind 1 of the door are full of zombies,
if you choose the wrong door, you will be eaten by them. Now, please make your choice.Get the 
2/3 chance to win
{'-'* 100} \n
    """)
    
    #using the random function to generate one random number from the selected area (1 to 3) 
    number = random.randint(1,3)  
    
    
    #using while loop and try / except function to check if the input number is correct
    while True:
        try:
            #asking player to enter the integer from one to three
            card_choice = int(input(prompt ="Please enter a number from 1 to 3: \n"))
            #conditional function for checking if the input is in the range
            #if not player will be asked to enter again 
            if card_choice < 0:
                print("There are no negative card!")
            elif card_choice > 3:
                print("There are only 3 numbers in the card!")
            #if the input equals to the random number
            elif card_choice == number:
                print(f"""Behind the door {number} are full of zombies.You are dead!""")
                fail()
            #if the input is not equals to the random number
            elif card_choice != number:
                print(f"""
                You win the game!! the door full of zombies is door {number}!
                You made the right choice!
                
                """)
                win()
            else:
                break
        #if the user input is not the number  
        except ValueError:
            print("Please enter in a correct format (in number)")
            
# using a UDF variable to win the game
def win():
    print(f"""

Congratulations!! You win the Game!

    """)
    
    input(prompt ='<Press enter to leave the game>\n')
    #using exit function to exit the game after winning 
    exit(0)
    
def fail():
    print(f"""

Game Over!!
Thanks for playing!!

    """)
    
    input(prompt ='<Press enter to leave the game>\n')
    #using exit function to exit the game after losing the game
    exit(0)
            
question()
```
